```python
import csv
import requests
import pandas as pd
from time import sleep
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def read_reports_from_csv():
    df = pd.read_csv("All_Scraped.csv")
    report_list = df["Report"].tolist()
    return report_list

previous_reports = read_reports_from_csv()
```

```python
options = Options()
# options.add_argument('--headless=new')
# options.add_argument('--ignore-certificate-errors')
driver = webdriver.Chrome(options=options)
driver.maximize_window()

url = 'https://galxe.com/spaces'

driver.get(url)
element = WebDriverWait(driver, 15).until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'div.space-box.position-relative')))
```

```python
def get_data(projects, new_scraped_writer, all_scraped_writer):
    for project in projects:
        if project not in previous_reports:
            data = {}

            driver.get(project)
            element = WebDriverWait(driver, 15).until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'div.sd-social-links')))

            try:
                data['Name'] = driver.find_element(By.CSS_SELECTOR, 'div.sd-title > span').text
            except:
                continue
            
            data['Report'] = project

            data['Website'] = ''
            
            data['Telegram'] = ''

            data['Twitter'] = ''

            data['Discord'] = ''

            new_scraped_writer.writerow(data)
            all_scraped_writer.writerow(data)
        else:
            continue
```

```python
def scroll_to_end():
    last_project = driver.find_elements(By.CSS_SELECTOR, 'div.space-box.position-relative')[-1]
    driver.execute_script("arguments[0].scrollIntoView(true);", last_project)

    time.sleep(3)
```

```python
header = ['Name','Website','Telegram','Twitter','Discord','Report']
with open("All_Scraped.csv", mode='a', newline='', encoding='utf-8') as all_scraped_file, open("New_Scraped.csv", mode='a', newline='', encoding='utf-8') as new_scraped_file:
    new_scraped_writer = csv.DictWriter(new_scraped_file, fieldnames=header)
    all_scraped_writer = csv.DictWriter(all_scraped_file, fieldnames=header)

    while True:
        old_page = driver.page_source
        scroll_to_end()
        new_page = driver.page_source

        # If there is no change in the page source, we have reached the end of the content
        if old_page == new_page:
            break
    
    projects = [project.get_attribute('href') for project in driver.find_elements(By.CSS_SELECTOR, 'div.space-box.position-relative > a')]
    get_data(projects, new_scraped_writer, all_scraped_writer)
```
